```python
import pandas as pd
```

```python
df = pd.read_csv('winequality-white.csv',delimiter=';')
```

```python
df.head()
```

```python
X=df.iloc[:,df.columns!='quality'].values
Y=df.iloc[:,df.columns=='quality'].values.flatten()
```

```python
from sklearn import preprocessing
#normalizing
normalized_X= preprocessing.normalize(X)
standardized_X = preprocessing.scale(X)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(standardized_X,Y,test_size=0.2,)
```

```python
from sklearn.linear_model import LinearRegression
lnrmodel = LinearRegression() 
```

```python
lnrmodel.fit(x_train,y_train)
```

```python
lnrmodel.score(x_test,y_test)
```

```python
prediction=lnrmodel.predict(x_test)
import numpy as np
predictions=np.round(prediction)
diff = predictions - y_test
correct=np.count_nonzero(diff==0)
score = correct/y_test.shape[0]
print("score after rounding regression results: ",score)
#MAD mean average deviation
numOfsample=y_test.shape[0]
sumofdeviation=abs(prediction-y_test).sum()
MAD=sumofdeviation/numOfsample
MAD
```
